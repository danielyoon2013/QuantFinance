```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
alphas = ['0.0','0.1','0.2','0.3']
acc_addition = []
for alpha in alphas:
    temp = pd.read_csv(f'result/additional_data_results_mix{alpha}.csv', index_col=0)
    temp = temp['Misclassification']
    temp.name=f'{float(alpha)*100}%'
    acc_addition.append(temp)
acc_addition = pd.concat(acc_addition, axis=1)
acc_addition
```

```python
plot_df = acc_addition
plt.figure()
for i in range(4):
    plt.plot(plot_df.iloc[i,:], label=plot_df.index[i])
plt.legend()
plt.title('Accuracy on Additional Test Data')
plt.ylabel('Accuracy on Test data')
plt.xlabel('Proportion of Addition Data in Train Set')
plt.ylim((0.4, 0.9))
plt.show()
```

```python
acc_origin = []
for alpha in alphas:
    temp = pd.read_csv(f'result/original_data_results_mix{alpha}.csv', index_col=0)
    temp = temp['Misclassification']
    temp.name=f'{float(alpha)*100}%'
    acc_origin.append(temp)
acc_origin = pd.concat(acc_origin, axis=1)
acc_origin
```

```python
plot_df = acc_origin
plt.figure()
for i in range(4):
    plt.plot(plot_df.iloc[i,:], label=plot_df.index[i])
plt.legend()
plt.title('Accuracy on Original Test Data')
plt.ylabel('Accuracy on Test data')
plt.xlabel('Proportion of Addition Data in Train Set')
plt.ylim((0.6, 0.9))
plt.show()
```
