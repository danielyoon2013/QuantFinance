```python
# Comparison

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

numbers = [10, 100, 500, 1000, 2000]
df = pd.DataFrame()
for n in numbers:
    s = "SVD_" + str(n) + "_components.csv"
    ndf = pd.read_csv(s, header=0)
    df = pd.concat( (df, ndf))
df['Model'] = df.iloc[:, 0]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()
for i in range(4):
    desired_model = df['Model'].unique()[i]
    df[df['Model'] == desired_model].plot(x='n_components', y=['AUC', 'Misclassification', 'MCC'], ax=axs[i])
    axs[i].set(title='Performance vs number of components, ' + desired_model,
              ylabel='Performance')
    
```
